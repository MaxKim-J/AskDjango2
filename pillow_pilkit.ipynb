```python
image_url = 'https://techcrunch.com/wp-content/uploads/2015/04/codecode.jpg?w=730&crop=1'
```

```python
import requests
```

```python
image_binary = requests.get(image_url).content
# cmd 기준 현재 디렉토리 경로로 다운로드
with open("flower.jpg", "wb") as f:
    f.write(image_binary)
```

```python
from PIL import Image
```

```python
flower_image = Image.open("flower.jpg")
```

```python
flower_image
```

```python
from pilkit.processors import Thumbnail
```

```python
processor = Thumbnail(width=300, height=300)
    # 프로세서 인스턴스 생성
    
thumb_image = processor.process(flower_image)
```

```python
thumb_image
```

```python
thumb_image.save("thumb-300x300.png")
    # 모드를 바꿔줘야 됨 투명 채널을 지원하지 않기 때문 
thumb_image.mode = "RGB"
    # 제이팩 압축률을 다르게 해서 저장하는 법
for quality in [100,80,60,40,20]:
    thumb_image.save("thumb-300x300--{}.jpg".format(quality), quality=quality)
```
